```python
from PIL import Image, ImageFilter
import numpy as np
from IPython.display import display

# 이미지 파일 경로 설정
image_path = '/content/drive/My Drive/Colab Notebooks/cat.jpg'

# 이미지 불러오기
image = Image.open(image_path)

# 이미지 크기 변경
new_width = 200
new_height = 300
resized_image = image.resize((new_width, new_height), Image.LANCZOS)

# 이미지 자르기
left = 50
upper = 50
right = 150
lower = 200
cropped_image = image.crop((left, upper, right, lower))


# 이미지 회전
degrees = 45
rotated_image = image.rotate(degrees)

# 이미지 상하, 좌우 대칭
flipped_image = image.transpose(Image.FLIP_LEFT_RIGHT)

# 이미지 필터링
blurred_image = image.filter(ImageFilter.BLUR)
contoured_image = image.filter(ImageFilter.CONTOUR)
detailed_image = image.filter(ImageFilter.DETAIL)
smoothed_image = image.filter(ImageFilter.SMOOTH_MORE)

# 이미지 저장
resized_image.save('/content/drive/My Drive/Colab Notebooks/resized_cat.jpg')

# 이미지를 바이트 배열로 변환
image_byte_array = resized_image.tobytes()

# 바이트 배열을 이미지로 변환하여 출력
image_from_bytes = Image.frombytes(resized_image.mode, resized_image.size, image_byte_array)

# 넘파이(numpy) 배열로부터 이미지 생성
image_from_np_array = Image.fromarray(np_array)

# 이미지 출력
display(resized_image, "Resized Image")
display(cropped_image, "Cropped Image")
display(rotated_image, "Rotated Image")
display(flipped_image, "Flipped Image")
display(blurred_image, "Blurred Image")
display(contoured_image, "Contoured Image")
display(detailed_image, "Detailed Image")
display(smoothed_image, "Smoothed Image")
display(image_from_bytes, "Image from Bytes")
display(image_from_np_array, "Image from NumPy Array")
```
